```python
from pathlib import Path
import pandas as pd
import utils
```

```python
# Example dataloading with a local toy catalogue (Python 3.9 compatible)
tmp = Path('tmp_data')
tmp.mkdir(exist_ok=True)

toy_members = pd.DataFrame({
    'uberID': [101, 102, 103, 104],
    'group_id_nessie': [1, 1, 2, -1],
})
toy_members_file = tmp / 'toy_members.csv'
toy_members.to_csv(toy_members_file, index=False)

loaded = utils.read_table(str(toy_members_file))
std = utils.standardise_membership(loaded, group_name='group_id_nessie')
std
```

```python
# Example of GAMA mapping aware membership loading
toy_cataid = pd.DataFrame({
    'CATAID': [1, 2, 3],
    'group_id': [10, 10, 11],
})
toy_map = pd.DataFrame({
    'CATAID': [1, 2, 3],
    'uberID': [1001, 1002, 1003],
})

members_file = tmp / 'toy_gama_members.csv'
map_file = tmp / 'toy_gama_map.csv'
toy_cataid.to_csv(members_file, index=False)
toy_map.to_csv(map_file, index=False)

mapped = utils.load_membership_with_optional_gama_mapping(
    str(members_file),
    output_group_name='group_id_nessie',
    gama_id_mapping_file=str(map_file),
)
mapped
```
